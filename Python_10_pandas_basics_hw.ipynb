```python
import pandas as pd
```

```python
ratings = pd.read_csv('ratings.csv')
```

```python
ratingsCount = ratings[ratings['rating']==5.0].groupby('movieId').agg(ratingCount=('rating', 'count'))
maxCount = ratingsCount['ratingCount'].max();
movideId = ratingsCount[ratingsCount['ratingCount']==maxCount]
print(movideId)
```

```python
power=pd.read_csv('power.csv')
```

```python
sum=0
filterPower=power[(power['year']>=2005)&(power['year']<=2010)&(power['quantity']>0)]
filterPowerCategory=filterPower[filterPower.category.isin([4,12,21])]
sum=filterPowerCategory[filterPowerCategory.country.str.contains('|'.join(['Lat','Lit','Est']))]['quantity'].sum()
print('Суммарное потребление стран Прибалтики:',sum)
```
